```python
from hyrax import Hyrax
import numpy as np

h = Hyrax()
```

```python
model_inputs = {
    "train": {
        "data": {
            "dataset_class": "HyraxRandomDataset",
            "data_location": "./data1_train",
            "primary_id_field": "object_id",
            "fields": ["image"],
            "dataset_config": {
                "shape": (3, 5, 5),
            },
        },
        "data_2": {
            "dataset_class": "HyraxRandomDataset",
            "data_location": "./data1_train",
            "primary_id_field": "object_id",
            "fields": ["image"],
            "dataset_config": {
                "shape": (3, 5, 5),
            },
        },
    },
    "infer": {
        "data": {
            "dataset_class": "HyraxRandomDataset",
            "data_location": "./data1_infer",
            "primary_id_field": "object_id",
            "dataset_config": {
                "shape": (3, 16, 16),
                "size": 10_000,
            },
        },
    },
}

h.config["model_inputs"] = model_inputs
```

```python
ds = h.prepare()
train_ds = ds["train"]
```

```python
from hyrax.pytorch_collate import hyrax_collate
```

```python
# This simulates the data loader requesting a batch size of 5.
sample = [train_ds[i] for i in range(5)]
```

```python
# Here we force a couple of the images to be different shapes than the rest to see the padding work.
# One is smaller, the other is larger than the rest.
sample[0]["data"]["image"] = np.ones((3, 3, 3))
sample[3]["data_2"]["image"] = np.ones((3, 6, 6))
```

```python
collated = hyrax_collate(sample)
```
