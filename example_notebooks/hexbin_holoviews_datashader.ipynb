# Hexbin Plot with HoloViews and Datashader

This notebook renders ~10 million 2D samples as a hexbin plot using HoloViews and Datashader. Datashader aggregates the data on the server side so the browser only receives a rasterized image — making it practical to visualize tens of millions of points interactively.

## 1. Import Required Libraries

```python
import numpy as np
import pandas as pd
import holoviews as hv
from holoviews.operation.datashader import rasterize
import datashader as ds

hv.extension("bokeh")
```

## 2. Generate Large-Scale 2D Sample Data

Generate ~150 million normally distributed 2D points (mean 0, std 1) and store them in a DataFrame.

```python
N = 150_000_000
rng = np.random.default_rng(seed=42)

df = pd.DataFrame(
    {
        "x": rng.normal(0, 1, N).astype(np.float32),
        "y": rng.normal(0, 1, N).astype(np.float32),
    }
)

print(f"Generated {len(df):,} points — memory: {df.memory_usage(deep=True).sum() / 1e6:.1f} MB")
```

## 3. Create a HoloViews HexTiles Element

`hv.HexTiles` is the correct element type for hexagonal bins. When `rasterize` receives a `HexTiles` element it routes to datashader's hexbin aggregator, producing Bokeh `hex_tile` glyphs. Using `hv.Points` would produce a rectangular pixel image instead.

```python
INITIAL_RANGE = (-3, 3)

tiles = hv.HexTiles(df, kdims=["x", "y"]).redim.range(x=INITIAL_RANGE, y=INITIAL_RANGE)
```

## 4. Adaptive Hexbin Aggregation

The callback recomputes `x_sampling`/`y_sampling` from the current viewport on every pan or zoom so that `TARGET_BINS` hexagons always span the visible x-axis — bins shrink as you zoom in and grow as you zoom out.

```python
TARGET_BINS = 50      # ~hexbins across x-axis at any zoom level
BUFFER_FACTOR = 0.2  # extra padding around viewport as a fraction of viewport width/height

range_xy = hv.streams.RangeXY()

def make_hexbin(x_range, y_range):
    # On initial render the stream hasn't fired yet — values are None.
    xr = x_range if (x_range is not None and None not in x_range) else INITIAL_RANGE
    yr = y_range if (y_range is not None and None not in y_range) else INITIAL_RANGE

    # Expand the filter region by BUFFER_FACTOR on each side so hexbins at the
    # edges of the viewport are computed using all their contributing points.
    x_pad = (xr[1] - xr[0]) * BUFFER_FACTOR
    y_pad = (yr[1] - yr[0]) * BUFFER_FACTOR
    mask = (
        (df["x"] >= xr[0] - x_pad) & (df["x"] <= xr[1] + x_pad) &
        (df["y"] >= yr[0] - y_pad) & (df["y"] <= yr[1] + y_pad)
    )
    df_view = df[mask]

    bin_size = (xr[1] - xr[0]) / TARGET_BINS

    # dynamic=False required: rasterize normally returns a DynamicMap, but nesting
    # a DynamicMap inside another DynamicMap is not supported.
    hexbin = rasterize(
        hv.HexTiles(df_view, kdims=["x", "y"]).redim.range(x=xr, y=yr),
        aggregator=ds.count(),
        x_sampling=bin_size,
        y_sampling=bin_size,
        dynamic=False,
    )

    # Apply log1p transform to the count values so the linear color mapper still
    # shows contrast across a wide dynamic range. log1p(x) = log(x+1) so bins
    # with count=1 map to log(2) > 0 and are always visible.
    # This avoids relying on Bokeh's LogColorMapper which has version compatibility issues.
    if hexbin.vdims:
        count_col = hexbin.vdims[0].name
        data = hexbin.data.copy()
        data[count_col] = np.log1p(data[count_col])
        hexbin = hexbin.clone(data)

    return hexbin

dmap = hv.DynamicMap(make_hexbin, streams=[range_xy])
```

## 5. Configure Plot Options

```python
plot = dmap.opts(
    hv.opts.HexTiles(
        cmap="Viridis",
        cnorm="linear",
        colorbar=True,
        colorbar_opts={"title": "log(Count + 1)"},
        width=1200,
        height=1100,
        xlabel="x",
        ylabel="y",
        title=f"Hexbin — {N:,} samples  |  ~{TARGET_BINS} bins across x",
        toolbar="above",
        tools=["hover", "pan", "wheel_zoom", "reset", "box_select", "lasso_select"],
        line_color="white",
        line_width=0.5,
        nonselection_alpha=1.0,   # prevent Bokeh from dimming unselected tiles on box-select,
        nonselection_line_alpha=1.0,  # keeping both tools visually consistent (outline only)
    )
)
```

## 6. Selection Streams

Two streams capture user selections from the plot:
- **`BoundsXY`** — fires with `(x0, y0, x1, y1)` after a box-select drag
- **`Lasso`** — fires with a polygon `geometry` after a lasso-select drag

Each drives a `DynamicMap` that draws the selection boundary as an overlay on the plot, and a Panel `Markdown` widget that shows the live point count inside the selection.

```python
import panel as pn
from matplotlib.path import Path as MplPath

pn.extension("tabulator")

# ── Selection streams ────────────────────────────────────────────────────────
bounds_stream = hv.streams.BoundsXY(source=dmap, bounds=(0, 0, 0, 0))
lasso_stream  = hv.streams.Lasso(source=dmap)

# Reactive text widgets — updated by the selection callbacks below
box_info   = pn.pane.Markdown("**Box selection:** none")
lasso_info = pn.pane.Markdown("**Lasso selection:** none")

# Selected subsets — accessible in downstream cells after making a selection
selected_box   = pd.DataFrame()
selected_lasso = pd.DataFrame()

# ── Selection table ──────────────────────────────────────────────────────────
MAX_TABLE_ROWS = 50_000   # cap rows rendered in the browser
_EMPTY = pd.DataFrame(columns=["row_index", "x", "y"])

selection_table = pn.widgets.Tabulator(
    _EMPTY.copy(),
    pagination="remote",
    page_size=25,
    show_index=False,
    sizing_mode="stretch_height",
    width=340,
    header_align="right",
    configuration={"columnDefaults": {"headerSort": True}},
    titles={"row_index": "Index", "x": "x", "y": "y"},
)

table_title = pn.pane.Markdown("### Selected Points", margin=(0, 0))

def _update_table(sel: pd.DataFrame) -> None:
    """Replace table contents with index + x,y columns of *sel*, capped at MAX_TABLE_ROWS."""
    if sel.empty:
        selection_table.value = _EMPTY.copy()
        table_title.object = "### Selected Points"
        return
    display = sel.iloc[:MAX_TABLE_ROWS][["x", "y"]].copy()
    display.insert(0, "row_index", display.index)
    display = display.reset_index(drop=True)
    selection_table.value = display
    table_title.object = f"### Selected Points — {len(sel):,}"

# Track previous stream values so the merged callback can detect which fired
_prev = {"bounds": (0, 0, 0, 0), "geometry": None}


def selection_overlay(bounds, geometry):
    """Single callback that handles both streams to avoid cross-stream cascade."""
    global selected_box, selected_lasso

    bounds_changed   = bounds   != _prev["bounds"]
    geometry_changed = geometry is not _prev["geometry"]
    _prev["bounds"]   = bounds
    _prev["geometry"] = geometry

    box_el   = hv.Rectangles([]).opts(apply_ranges=False)
    lasso_el = hv.Polygons([]).opts(apply_ranges=False)

    if bounds_changed:
        # ── Box selection ────────────────────────────────────────────────────
        selected_lasso = pd.DataFrame()

        x0, y0, x1, y1 = bounds
        if x0 != x1 and y0 != y1:
            mask = (
                (df["x"] >= x0) & (df["x"] <= x1) &
                (df["y"] >= y0) & (df["y"] <= y1)
            )
            selected_box = df[mask]
            _update_table(selected_box)
            box_el = hv.Rectangles([(x0, y0, x1, y1)]).opts(
                fill_alpha=0.1, fill_color="cyan", line_color="cyan",
                line_width=1.5, apply_ranges=False,
            )
        else:
            selected_box = pd.DataFrame()
            _update_table(pd.DataFrame())

    elif geometry_changed and geometry is not None:
        # ── Lasso selection ──────────────────────────────────────────────────
        selected_box = pd.DataFrame()

        if isinstance(geometry, dict):
            coords = np.array(geometry["coordinates"][0])
        else:
            coords = np.asarray(geometry)

        if len(coords) >= 3:
            # Pre-filter to bounding box before the expensive point-in-polygon test
            x_min, y_min = coords[:, 0].min(), coords[:, 1].min()
            x_max, y_max = coords[:, 0].max(), coords[:, 1].max()
            bbox_mask = (
                (df["x"] >= x_min) & (df["x"] <= x_max) &
                (df["y"] >= y_min) & (df["y"] <= y_max)
            )
            df_candidates = df[bbox_mask]

            path = MplPath(coords)
            inside = path.contains_points(df_candidates[["x", "y"]].values)

            selected_lasso = df_candidates[inside]
            _update_table(selected_lasso)
            lasso_el = hv.Polygons([{("x", "y"): coords}]).opts(
                fill_alpha=0.1, fill_color="orange", line_color="orange",
                line_width=1.5, apply_ranges=False,
            )
        else:
            selected_lasso = pd.DataFrame()
            _update_table(pd.DataFrame())

    return box_el * lasso_el



selection_dmap = hv.DynamicMap(selection_overlay, streams=[bounds_stream, lasso_stream])
```

## 7. Display

Overlay the selection shapes on the hexbin plot and show the live point-count widgets above it. Switch between **Box Select** and **Lasso Select** using the toolbar buttons.

```python
combined = (plot * selection_dmap).opts(
    hv.opts.Rectangles(apply_ranges=False),
    hv.opts.Polygons(apply_ranges=False),
)

# ── Export button ────────────────────────────────────────────────────────────
_ipy = get_ipython()

export_btn    = pn.widgets.Button(name="Export to selected_points", button_type="primary", width=340)
export_status = pn.pane.Markdown("", width=300)

def _on_export(event):
    sel = selected_box if not selected_box.empty else selected_lasso
    if sel.empty:
        export_status.object = "_No selection to export._"
        return
    _ipy.user_ns["selected_points"] = sel
    export_status.object = f"_Exported **{len(sel):,}** rows to `selected_points`._"

export_btn.on_click(_on_export)

# ── Theme toggle ─────────────────────────────────────────────────────────────
from bokeh.themes import built_in_themes, Theme as BokehTheme

theme_toggle = pn.widgets.Switch(value=False)

def _on_theme(event):
    # Update the live Bokeh document so existing glyphs/axes repaint immediately
    doc = pn.state.curdoc
    if doc is not None:
        doc.theme = built_in_themes["dark_minimal"] if event.new else BokehTheme(json={})
    # Update the HoloViews renderer so future re-renders use the same theme
    hv.renderer("bokeh").theme = "dark_minimal" if event.new else None
    # Re-fire the range stream to trigger a hexbin re-render under the new theme
    xr = range_xy.x_range or INITIAL_RANGE
    yr = range_xy.y_range or INITIAL_RANGE
    range_xy.event(x_range=xr, y_range=yr)

theme_toggle.param.watch(_on_theme, "value")

# ── Layout ───────────────────────────────────────────────────────────────────
table_header = pn.Row(
    table_title,
    pn.HSpacer(),
    pn.pane.Markdown("☀️", margin=(0, 2)),
    theme_toggle,
    pn.pane.Markdown("🌙", margin=(0, 2)),
    align="center",
    width=340,
)

pn.Column(
    pn.Row(
        combined,
        pn.Column(
            table_header,
            selection_table,
            pn.Spacer(height=10),
            export_btn,
            export_status,
            sizing_mode="stretch_height",
        ),
    ),
).servable()
```
