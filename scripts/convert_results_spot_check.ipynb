# Manual check for conversion

This notebook stub is here to help ensure that the conversion script did what was expected.
Namely convert older style .npy output to .lance dataset

```python
# Replace these with the approprate paths on your system

original_dataset_directory = "/home/drew/code/hyrax/docs/pre_executed/results/old_npy_cnn_output"
converted_dataset_directory = "/home/drew/code/hyrax/docs/pre_executed/results/old_npy_cnn_output"
```

## Create Hyrax dataset instances
Create an instance of the original dataset and the converted dataset for comparison

```python
from hyrax.datasets.inference_dataset import InferenceDataset

inference_dataset = InferenceDataset({}, original_dataset_directory)
```

```python
from hyrax.datasets.result_dataset import ResultDataset

results_dataset = ResultDataset({}, converted_dataset_directory)
```

## Simple length check

```python
assert len(inference_dataset) == len(results_dataset)
```

## Verify the ids
Check to ensure that the ids of both old and new dataset are the same

```python
inference_ids = list(inference_dataset.ids())
results_ids = list(results_dataset.ids())

assert inference_ids == results_ids
```

## Spot check values
Here we just check that the result data values are the same.
Note that if a particularly old Inference dataset is being converted, the original my be a pytorch datatype.
When converting to a ResultDataset it will have been changed into a numpy datatype before being saved.
There should be no loss of precision, but the default representation of values differs between InferenceDataset and ResultDataset.

We can see that when we convert them both to numpy float64 arrays, the values are the same.

```python
import numpy as np

print("Raw InferenceResult data:")
print(inference_dataset[0])

print("Converted to numpy float64:")
converted_inferece_sample = np.asarray(inference_dataset[0], dtype=np.float64)
print(converted_inferece_sample)
```

```python
print("Raw ResultDataset data:")
print(results_dataset[0])

print("Converted to numpy float64:")
converted_result_sample = np.asarray(results_dataset[0], dtype=np.float64)
print(converted_result_sample)
```

```python
assert np.array_equal(converted_inferece_sample, converted_result_sample)
```
