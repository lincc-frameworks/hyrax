# MMU Spectra with Hyrax

This notebook shows how to configure Hyrax to read **spectral** data from Multimodal Universe using `MultimodalUniverseDataset`.


```python
import hyrax

h = hyrax.Hyrax()

h.config["data_request"] = {
    "infer": {
        "mmu": {
            "dataset_class": "MultimodalUniverseDataset",
            "data_location": "hf://MultimodalUniverse/desi",
            "primary_id_field": "object_id",
            # Filter fields to just what you need
            # "fields": ["spectrum"],
            "dataset_config": {
                "MultimodalUniverseDataset": {
                    "split": "train",
                    "max_samples": 100,
                    "streaming": True,
                }
            },
        }
    }
}

prepared = h.prepare()
print(f"Loaded rows: {len(prepared['infer'])}")
```

After loading, inspect `prepared['infer'][0].keys()` and set `fields` to the exact MMU columns you want.


```python
prepared["infer"][0]["mmu"].keys()
```

```python
prepared["infer"][0]["mmu"]["spectrum"].keys()
```

```python
for key, value in prepared["infer"][0]["mmu"]["spectrum"].items():
    print(f"{key}: {type(value)} {value.shape} {value.dtype}")
```

```python
import matplotlib.pyplot as plt

wavelength = prepared["infer"][0]["mmu"]["spectrum"]["lambda"]
flux = prepared["infer"][0]["mmu"]["spectrum"]["flux"]

plt.figure(figsize=(10, 5))
plt.step(wavelength, flux, where="mid", color="black", linewidth=0.5)
plt.title("Example Spectrum")
plt.xlabel("Wavelength ($\AA$)")
plt.ylabel("Flux ($10^{-17} \\text{erg}\\ s^{-1} \\text{cm}^{-2} \AA^{-1}$)")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```
