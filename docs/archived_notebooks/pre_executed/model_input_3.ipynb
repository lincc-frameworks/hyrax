# Providing Data - Level 3
## Dataset splits with split_fraction
Dataset splits are configured using `split_fraction` in `data_request` groups. Groups pointing to the same `data_location` with `split_fraction` values receive non-overlapping subsets of the data.

## Defining what data to use for inference
Introduce `data_request.infer` as the way to define what data is provided to the model during inference.
