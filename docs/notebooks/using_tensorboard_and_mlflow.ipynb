# Using TensorBoard and MLFlow

This notebook shows how to activate a server to use TensorBoard and MLFlow.

## TensorBoard from a Notebook

To view TensorBoard in a notebook, add a code block with the following two lines:
```bash
%load_ext tensorboard
%tensorboard --logdir results/
```
This will load TensorBoard directly in the notebook. To reload TensorBoard without restarting the notebook, use the following instead:
```bash
%reload_ext tensorboard
%tensorboard --logdir results/
```

## TensorBoard from the Terminal

Whenever training or inference is completed, a `results` directory is generated that contains timestamped "train" and "infer" folders

![results_dir.png](results_dir.png)

To activate TensorBoard, navigate to the `results` directory from the terminal and activate a virtual environment that has TensorBoard installed. From this directory, run `tensorboard --logdir .`. In general, it is acceptable to run `tensorboard --logdir {path_to_results_directory}`.

TensorBoard should now be active and the last line on the terminal should read `TensorBoard 2.20.0 at http://localhost:6006/ (Press CTRL+C to quit)`. Navigate to `http://localhost:16006/` from the URL bar of any web browser to use TensorBoard.

## MLFlow from a Notebook

To view MLFlow in a notebook, execute a Python code block with the following code

```Python
import subprocess
from IPython.display import IFrame

backend_store_uri = f"./{path_to_results_directory}/mlflow"
mlflow_ui_process = subprocess.Popen(
    ["mlflow", "ui", "--backend-store-uri", backend_store_uri, "--port", "8080"],
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE,
)

IFrame(src="http://localhost:8080", width="100%", height=1000)
```

## MLFlow from the Terminal

To launch MLFlow from the terminal, simply run `mlflow server --port 6006 --backend-store-uri file:///{path_to_results_directory}/mlflow/`. MLFlow can be accessed from any web browser at `http://localhost:6006/`.

## A Note on Port Forwarding

If the `results` directory is stored on a remote server that is `ssh`ed into, the `ssh` command has a `-L` option for port forwarding. For example, running `ssh -L 16006:127.0.0.1:6006 {user@remote}` will connect the local port `16006` to the remote port `6006`.

This means TensorBoard and MLFlow can be activated from the remote `results` directory in the remote server's terminal and accessed from a local web browser at `http://localhost:6006`.

To view TensorBoard on a notebook on the remote server, no port forwarding is required. For MLFlow, the `-L` cooption is necessary
