# Hyrax HATS Dataset Example

Minimal Phase-1 example using `HyraxHATSDataset` with an LSDB-written HATS catalog.

```python
from pathlib import Path
import tempfile

import pandas as pd
import lsdb
import hyrax
```

```python
tmp_dir = Path(tempfile.mkdtemp())
catalog_path = tmp_dir / "sample_hats"
df = pd.DataFrame(
    {
        "object_id": [1001, 1002, 1003],
        "coord_ra": [150.1, 150.2, 150.3],
        "coord_dec": [2.1, 2.2, 2.3],
        "mag-r": [19.1, 19.4, 18.9],
    }
)
lsdb.from_dataframe(df, ra_column="coord_ra", dec_column="coord_dec").write_catalog(catalog_path)
catalog_path
```

```python
h = hyrax.Hyrax()
h.config["data_request"] = {
    "train": {
        "data": {
            "dataset_class": "HyraxHATSDataset",
            "data_location": str(catalog_path),
            "fields": ["coord_ra", "mag-r"],
            "primary_id_field": "object_id",
        }
    }
}
dataset = h.prepare()["train"]
len(dataset)
```
